```python
import pandas as pd
import jieba
```

```python
x_train, y_train = [], []
with open('./data/cnews.train.txt', 'r', encoding = 'utf8') as tr:
    for line in tr:
        label, content = line.strip().split('\t')
        if content:
            x_train.append(content)
            y_train.append(label)

print(x_train[:10])
print(y_train[:10])
print(len(x_train))
print(len(y_train))
```

```python
"""读取分类目录，固定"""
categories = ['体育', '财经', '房产', '家居', '教育', '科技', '时尚', '时政', '游戏', '娱乐']
cat_to_id = dict(zip(categories, range(len(categories))))
cat_to_id
```

```python
y_train = [cat_to_id[_] for _ in y_train]
```

```python
# 载入停用词库
stopwords_dict = {}
stopwords = open('../sklearn-tfidf/chinese_stopwords.txt').readlines()
for i in stopwords:
    stopword = i.rstrip()
    if stopword == '':
        stopword = ' '
    stopwords_dict[stopword] = True
print('呀' in stopwords_dict, "'呀'是停用词")
print('福尔摩斯' in stopwords_dict, "'福尔摩斯'不是停用词")
def process(word):
    stopword = word.rstrip()
    if stopword == '':
        stopword = ' '
    return stopword
stopwords = [process(word) for word in stopwords] 
```

```python
x_train = [' '.join(jieba.cut(content)) for content in x_train]
x_train[:10]
```

```python
x_test, y_test = [], []
with open('./data/cnews.val.txt', 'r', encoding = 'utf8') as tr:
    for line in tr:
        label, content = line.strip().split('\t')
        if content:
            x_test.append(content)
            y_test.append(label)
y_test = [cat_to_id[_] for _ in y_test]
x_test = [' '.join(jieba.cut(content)) for content in x_test]
len(x_test)
```

```python
x_test[1]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# 由于已经中文分词了，模型使用ngram为1的结构，停用词使用引入的词库
tfv = TfidfVectorizer(ngram_range=(1, 1),stop_words=stopwords,use_idf=1,smooth_idf=1,sublinear_tf=1)

# 在抽取关键词时，train_tfidf是原始词频将其中的内容作为语料库
train_tfidf = tfv.fit_transform(x_train)
words = tfv.get_feature_names()
test_tfidf = tfv.transform(x_test)
print("关键词抽取完成")
```

```python
# 多项式朴素贝叶斯
from sklearn.naive_bayes import MultinomialNB as MNB
from sklearn.naive_bayes import BernoulliNB as BNB
model_NB = MNB()
model_NB.fit(train_tfidf, y_train) #特征数据直接灌进来
MNB(alpha=1.0, class_prior=None, fit_prior=True)
```

```python
from sklearn.cross_validation import cross_val_score
import numpy as np
print("多项式朴素贝叶斯分类器20折交叉验证得分: ", np.mean(cross_val_score(model_NB, train_tfidf, y_train, cv=20, scoring='accuracy')))
```

```python
predict = model_NB.predict(test_tfidf)
print("测试集数量", len(predict))
# 判断正确率并打印出判断出错的结果。
def compare(predict, test):
    sum = len(predict)
    great = 0
    for i in range(sum):
        if (predict[i] == test[i]):
            great += 1
#         else:
#             print(test_dict[i],"\n", test_Y[i] and "句子有问题\n" or "句子没问题\n")
    print("测试数据的正确率: ", great / sum)
compare(predict, y_test)
```
