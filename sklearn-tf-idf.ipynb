```python
import os
import jieba
import numpy as np
titlelist = []
root = "福尔摩斯探案大全集"
# 先读取标题序列，遍历文件夹
for file in os.listdir(root):
    if '.' not in root:
        # 遍历文档
        for content in os.listdir(root + "/" + file):
            # 标题
            titlelist.append(file + '--' + content.split('.')[0].replace("\u3000", " "))
print(titlelist)
```

```python
# 直观来看，先读取一个文档
with open("福尔摩斯探案大全集//冒险史系列//五个桔核.txt", 'r',encoding = 'utf-8') as f:
    text = f.read().strip().replace('\n', '').replace(' ', '').replace('　', '').replace('\t', '').replace('\r', '')
# print(content)
```

```python
# 载入停用词库
stopwords_dict = {}
stopwords = open('chinese_stopwords.txt').readlines()
for i in stopwords:
    stopword = i.rstrip()
    if stopword == '':
        stopword = ' '
    stopwords_dict[stopword] = True
print('呀' in stopwords_dict)
print('福尔摩斯' in stopwords_dict)
def process(word):
    stopword = word.rstrip()
    if stopword == '':
        stopword = ' '
    return stopword
stopwords = [process(word) for word in stopwords] 
stopwords
```

```python
# 分词
# 增加分词词库，使分词更准确
jieba.add_word(u'劳瑞斯顿')
jieba.add_word(u'雷斯垂德')
jieba.add_word(u'葛莱森')
seg_list = jieba.cut(text)
# seg_list = jieba.cut(content, cut_all=True)

# 过滤停用词
# seg_list_after = []
# for item in seg_list:
#     if item and item not in stopwords_dict:
#         seg_list_after.append(item)
result = ' '.join(seg_list)
print(result)
```

<a href="http://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html#sklearn.feature_extraction.text.TfidfVectorizer">点击查看官方sklearn TFIDF向量类</a>

TfidfVectorizer:
- --ngram_range: 元组，设定ngram的范围，这里用1~3元ngram
- --stop_words: 填'english'的话，使用默认英语停用词，也可以是一个List（停用词序列），默认是None（不使用停用词）
- --use_idf : 启用逆文档频率重新加权。默认True
- --smooth_idf : 通过在文档频率上加一个平滑的idf权重，相当于有一个额外的文档包含语料库中的每一个词一次。防止零分割。默认True
- --sublinear_tf : 应用次线性tf缩放，即用1 + log（tf）替换tf。默认False

```python
# 训练Tfidf模型
# from sklearn.feature_extraction.text import CountVectorizer
# from sklearn.feature_extraction.text import TfidfTransformer
# vectorizer = CountVectorizer()
# word_frequence = vectorizer.fit_transform([result])
# words = vectorizer.get_feature_names()
# transformer = TfidfTransformer()
# tfidf = transformer.fit_transform(word_frequence)
# weight = tfidf.toarray()
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(ngram_range=(1, 1),stop_words=stopwords,use_idf=1,smooth_idf=1,sublinear_tf=1)
word_frequence = tfv.fit_transform([result])
words = tfv.get_feature_names()
weight = word_frequence.toarray()
```

```python
for (title, w) in zip(titlelist, weight):
    print(u'{}:'.format(title))
    # 排序
    loc = np.argsort(-w)
    # 显示前20个高词频词语。（现在只有一个文档，tfidf的结果只是词频）
    # 可以看到"福尔摩斯"是词频最高的，但其实"福尔摩斯"并不是这篇文章的关键词。
    for i in range(20):
        print(u'-{}: {} {}'.format(str(i + 1), words[loc[i]], w[loc[i]]))
    print('\n')
```

```python
# 现在读取并统计所有文档
wordlist = []
for file in os.listdir(root):
    if '.' not in root:
        # 遍历文档
        for content in os.listdir(root + "/" + file):
            with open(root + "//" + file + "//" + content, 'r',encoding = 'utf-8') as f:
                text = f.read().strip().replace('\n', '').replace(' ', '').replace('　', '').replace('\t', '').replace('\r', '')
                # 分词
                seg_list = jieba.cut(text)
                # seg_list = jieba.cut(content, cut_all=True)
                result = ' '.join(seg_list)
                wordlist.append(result)
            print(root + "//" + file + "//" + content + " 载入完成")
```

```python
tfv = TfidfVectorizer(ngram_range=(1, 1),stop_words=stopwords,use_idf=1,smooth_idf=1,sublinear_tf=1)
word_frequence = tfv.fit_transform(wordlist)
words = tfv.get_feature_names()
weight = word_frequence.toarray()
print("关键词抽取完成")
```

```python
# 输出每篇文章靠前的关键词，在这里的人名分词问题，可以通过增加分词词库来解决。
n = 20
for (title, w) in zip(titlelist, weight):
    print(u'{}:'.format(title))
    # 排序
    loc = np.argsort(-w)
    for i in range(n):
        print(u'-{}: {} {}'.format(str(i + 1), words[loc[i]], w[loc[i]]))
    print('\n')
# 以《五个桔核》为例，这里的"桔核"所占tfidf权重最高，说明"桔核"是一个关键词。
# 而词频最高的"福尔摩斯"在多个文件中有很多复现，因而在这篇文章中的关键程度降低了。
```
